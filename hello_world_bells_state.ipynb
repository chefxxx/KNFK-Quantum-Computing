```python
from qiskit import *
```

```python
cir = QuantumCircuit(2,2)
```

```python
cir.draw(output='mpl')
```

```python
cir.h(0)
```

```python
cir.cx(0, 1)
```

```python
cir.measure([0, 1], [0, 1])
```

```python
cir.draw(output='mpl')
```

```python
from qiskit_aer import AerSimulator
```

```python
backend = AerSimulator()
```

```python
qc_compiled = transpile(cir, backend)
```

```python
job_sim = backend.run(qc_compiled, shots=2048)
```

```python
result_sim = job_sim.result()
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(result_sim.get_counts(qc_compiled))
```

```python
from qiskit_ibm_provider import IBMProvider, least_busy
```

```python
provider = IBMProvider()
```

```python
least_busy(provider.backends())
```

```python
qcomp = provider.get_backend('ibmq_qasm_simulator')
```

```python
job = qcomp.run(cir)
```

```python
from qiskit.tools.monitor import job_monitor
```

```python
job_monitor(job)
```

```python
result = job.result()
```

```python
plot_histogram(result.get_counts())
```
